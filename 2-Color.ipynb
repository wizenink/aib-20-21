# Color

Por defecto en OpenCV las imágenes en color se cargan en formato BGR: el primer canal es el canal azul (B), el segundo canal es el verde (G) y el tercer canal, el rojo (R).

Para cambiar de espacio de color existe la función `cvtColor` que recibe como primer parámetro la imagen a transformar y, como segundo parámetro, una constante que representa el espacio de color actual de la imagen y el espacio de color de salida de la imagen. La salida de esta función es la imagen transformada al espacio de color elegido.

```python
%matplotlib inline
from matplotlib import pyplot as plt

import numpy as np
import cv2


im = cv2.imread('res/butterfly.jpg')

im_rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)

plt.imshow(im_rgb)
plt.show()
```

Algunos de los posibles de cambios de espacio color son:

- GRAY: `cv2.COLOR_BGR2GRAY`, `cv2.COLOR_RGB2GRAY`, `COLOR_GRAY2BGR, COLOR_GRAY2RGB`
- XYZ: `cv2.COLOR_BGR2XYZ`, `cv2.COLOR_RBG2XYZ`, `cv2.COLOR_XYZ2RGB`, `cv2.COLOR_XYZ2BGR`, 
- YCrCb: `cv2.COLOR_BGR2YCrCb`, `cv2.COLOR_RBG2YCrCb`, `cv2.COLOR_YCrCb2BGR`, `cv2.COLOR_YCrCb2RGB`
- HSV: `cv2.COLOR_BGR2HSV`, `cv2.COLOR_RGB2HSV`, `cv2.COLOR_HSV2BGR`, `cv2.COLOR_HSV2RGB`
- HSL: `cv2.COLOR_BGR2HLS`, `cv2.COLOR_RGB2HLS`, `cv2.COLOR_HLS2BGR`, `cv2.COLOR_HLS2RGB`
- Lab: `cv2.COLOR_BGR2Lab`, `cv2.COLOR_RGB2Lab`, `cv2.COLOR_Lab2BGR`, `cv2.COLOR_Lab2RGB`
- Luv: `cv2.COLOR_BGR2Luv`, `cv2.COLOR_RGB2Luv`, `cv2.COLOR_Luv2BGR`, `cv2.COLOR_Luv2RGB`

1) Convierte la imagen de la mariposa de BGR a los espacios de color RGB, HSV y Lab. En cada uno de estos espacios de color:
a) Visualiza cada uno de los canales por separado
b) Comprueba cuáles son los valores mínimos y máximos de cada canal


```python
# Escribe aquí tu código

rgbfig,rgbax = plt.subplots(2,2,sharex=True,sharey=True)
rgb = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
rgbax[0,0].imshow(rgb)
rgbax[0,0].set_title("RGB")
rgbax[0,1].imshow(rgb[:,:,0],cmap='gray')
rgbax[0,1].set_title("{}(min:{},max:{})".format("R",np.amin(rgb[:,:,0]),np.amax(rgb[:,:,0])))
rgbax[1,0].imshow(rgb[:,:,1],cmap='gray')
rgbax[1,0].set_title("{}(min:{},max:{})".format("G",np.amin(rgb[:,:,1]),np.amax(rgb[:,:,1])))
rgbax[1,1].imshow(rgb[:,:,2],cmap='gray')
rgbax[1,1].set_title("{}(min:{},max:{})".format("B",np.amin(rgb[:,:,2]),np.amax(rgb[:,:,2])))
rgbfig.suptitle("RGB Color Space")


hsvfig,hsvax = plt.subplots(2,2,sharex=True,sharey=True)
hsv = cv2.cvtColor(im,cv2.COLOR_BGR2HSV)
hsvax[0,0].imshow(hsv)
hsvax[0,0].set_title("HSV as RGB")
hsvax[0,1].imshow(hsv[:,:,0],cmap='gray')
hsvax[0,1].set_title("{}(min:{},max:{})".format("H",np.amin(hsv[:,:,0]),np.amax(hsv[:,:,0])))
hsvax[1,0].imshow(hsv[:,:,1],cmap='gray')
hsvax[1,0].set_title("{}(min:{},max:{})".format("S",np.amin(hsv[:,:,1]),np.amax(hsv[:,:,1])))
hsvax[1,1].imshow(hsv[:,:,2],cmap='gray')
hsvax[1,1].set_title("{}(min:{},max:{})".format("V",np.amin(hsv[:,:,2]),np.amax(hsv[:,:,2])))
hsvfig.suptitle("HSV Color Space")

labfig,labax = plt.subplots(2,2,sharex=True,sharey=True)
lab = cv2.cvtColor(im,cv2.COLOR_BGR2LAB)
labax[0,0].imshow(lab)
labax[0,0].set_title("LAB as RGB")
labax[0,1].imshow(lab[:,:,0],cmap='gray')
labax[0,1].set_title("{}(min:{},max:{})".format("L",np.amin(lab[:,:,0]),np.amax(lab[:,:,0])))
labax[1,0].imshow(lab[:,:,1],cmap='gray')
labax[1,0].set_title("{}(min:{},max:{})".format("A",np.amin(lab[:,:,1]),np.amax(lab[:,:,1])))
labax[1,1].imshow(lab[:,:,2],cmap='gray')
labax[1,1].set_title("{}(min:{},max:{})".format("B",np.amin(lab[:,:,2]),np.amax(lab[:,:,2])))
labfig.suptitle("LAB Color Space")

```
